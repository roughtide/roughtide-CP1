```python
import pandas as pd
import datetime

schedule = pd.read_csv('C:\\Develop\\Project7\\dataproc\\schedule2022.csv', usecols=['date', 'weekday', 'ranking'])
schedule.date = pd.to_datetime(schedule.date)
days = schedule.date.dt.strftime('%Y%m%d').values
days
```

```python
import pandas as pd
import json
import requests
from bs4 import BeautifulSoup

url = 'http://apis.data.go.kr/1360000/AsosDalyInfoService/getWthrDataList'

df = pd.DataFrame(columns=['stnId','tm','avgTa','sumRn'])
for day in days:
    queryParams = '?' + \
                'ServiceKey=' + 'YeSiYlUdLxSa9aOaPFM1kADTX45LzrMBd%2F43zl9fzjusryc9kIaewAPe9FmIZU5h09iiB4SFHX1UKe9jLQcSRw%3D%3D'+ \
                '&pageNo='+ '1' + \
                '&numOfRows='+ '999' + \
                '&dataType='+ 'JSON' + \
                '&dataCd='+ 'ASOS' + \
                '&dateCd='+ 'DAY' + \
                '&startDt='+ day + \
                '&endDt='+ day + \
                '&stnIds='+ '112'

    result = requests.get(url + queryParams)
    js = json.loads(result.content)
    data = pd.DataFrame(js['response']['body']['items']['item'])
    li = ['stnId','tm','avgTa','sumRn']
    data.loc[:,li]
    df = pd.concat((df,data.loc[:,li]),sort=False)
    
```

```python
df = df[['tm', 'avgTa', 'sumRn']]
```

```python
df.sumRn = df.sumRn.replace('',0)
df = df.rename(columns={'tm':'date','avgTa':'avgtemp','sumRn':'rainfall'})
df
```

```python
df.to_csv('dataproc\\weather.csv')
```
