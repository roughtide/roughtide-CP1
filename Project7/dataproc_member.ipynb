```python
ziptable = zipcode[zipcode.country_code == 'KR'][['postal_code','admin_name2', 'latitude','longitude']]
```

```python
ziptable.rename(columns={'postal_code':'zipcode'}, inplace=True)
ziptable
```

```python
ziptable.zipcode = ziptable.zipcode.astype(int) # 0으로 시작하는 
ziptable.zipcode = ziptable.zipcode.astype(str)
```

```python
member = pd.merge(member, ziptable, how='left', on='zipcode')
```

인천유나이티드경기장의 위도37.5685, 경도126.9469

```python
from haversine import haversine
```

```python
member['dist'] = 0
```

```python
stadium = (37.466047, 126.643011)
for i in range(len(member)):
    point = (member.latitude[i], member.longitude[i])
    member['dist'][i] = haversine(stadium, point, unit = 'km')
```

```python
member[member.dist > 300] # 거주지로부터 300km 이상인 고객들
```
