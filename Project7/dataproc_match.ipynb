```python
parser = etree.XMLParser(encoding="euc-kr")
tree = etree.parse('C:\\Develop\\Project7\\dataset\\schedule2022120221019212245.xml', parser=parser)


pageIds = tree.find('row')
schedule = pd.DataFrame(columns = [x[0] for x in pageIds.items()])
pageIds = tree.findall('row')
for page in pageIds:
    mydict = dict(page.items())
    condf = pd.DataFrame.from_dict(mydict, orient='index').T
    schedule = pd.concat([schedule, condf], ignore_index=True)

schedule
```

```python
schedule = schedule[(schedule['home_team_name']=='인천') | (schedule['away_team_name']=='인천')] # 클라이언트의 경기일정만 추출
schedule = schedule.dropna() # 관중수 0인 결측데이터로 현 경기결과를 찾아보면서까지 활용할 가치가 없음
schedule
```

```python
schedule['home'] = schedule['home_team_name'] == '인천' # home adv
schedule.reset_index(drop=True, inplace=True)
```

```python
opp = []
for i in range(len(schedule)):
    if schedule.home.loc[i] == True:
        opp.append(schedule.away_team_name.loc[i])
    else:
        opp.append(schedule.home_team_name.loc[i])
schedule['opponent'] = pd.Series(data=opp)  # vs team
```

```python
goal_incheon = []
goal_diff = []
for i in range(len(schedule)):
    if schedule.home.loc[i] == True:
        goal_incheon.append(schedule.home_team_goal.loc[i])
        goal_diff.append(int(schedule.home_team_goal.loc[i]) - int(schedule.away_team_goal.loc[i]))
    else:
        goal_incheon.append(schedule.away_team_goal.loc[i])
        goal_diff.append(int(schedule.away_team_goal.loc[i]) - int(schedule.home_team_goal.loc[i]))
schedule['goal_incheon'] = pd.Series(data=goal_incheon)  # 인천 득점
schedule['goal_diff'] = pd.Series(data=goal_diff)  # 상대와의 득점차(양수 = 승리, 음수 = 패배)
```

```python
import datetime
import requests
from bs4 import BeautifulSoup, element
```

```python
def getHolidayDataFrame(year: int) -> pd.DataFrame:
    url = 'http://apis.data.go.kr/B090041/openapi/service/SpcdeInfoService/getRestDeInfo'
    api_key = 'YeSiYlUdLxSa9aOaPFM1kADTX45LzrMBd%2F43zl9fzjusryc9kIaewAPe9FmIZU5h09iiB4SFHX1UKe9jLQcSRw%3D%3D'
    api_key_decoded = requests.utils.unquote(api_key, encoding='utf-8')

    params = {
        'serviceKey': api_key_decoded,
        'solYear': year,
        'numOfRows': 100
    }

    weekday_name = ['월','화','수','목','금','토','일']

    res = requests.get(url,params=params)
    xml = BeautifulSoup(res.text, features="lxml")

    def parseItem(item: element.Tag) -> dict:
        dt = datetime.datetime.strptime(item.find('locdate').text.strip(), '%Y%m%d')
        item_dict = {
            'name': item.find('datename').text.strip(),
            'date': dt,
            'weekday': weekday_name[dt.weekday()]
        }
        return item_dict

    item_list = [parseItem(x) for x in xml.find('items')]
    return pd.DataFrame(item_list)
```
