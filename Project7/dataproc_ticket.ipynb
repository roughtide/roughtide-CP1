```python
ticket = ticket.drop(columns=['match_info', 'reserve_state_code', 'reserve_detail_no', 'ticket_no', 'seat_Attribute_4', 'seat_Attribute_5', 'seat_Attribute_6', 'detail_cancel_date_time'])
```

```python
ticket.head(30)
```

```python
ticket['reserve_datetime'] = pd.to_datetime(ticket['reserve_datetime'])
ticket['cancel_datetime'] = pd.to_datetime(ticket['cancel_datetime'])
```

```python
ticket.game_date = ticket.game_date.astype(str)
```

```python

ticket['timeremain'] = ticket.game_date + ' ' + ticket.game_time
ticket['timeremain'] = pd.to_datetime(ticket['timeremain'])
```

```python
ticket.cancel_datetime = round((ticket.timeremain - ticket.cancel_datetime).dt.total_seconds()/60)
ticket.reserve_datetime = round((ticket.timeremain - ticket.reserve_datetime).dt.total_seconds()/60)
```
